```python
#df['assignment']= df['assignment'].astype("int")

# Replace dictionary
replace_dict = {
    1: "text_only",
    2: "audio_only",
    3: "video_only",
    4: "audio+text",
    5: "video+text",
    6: "video+audio",
    7: "video+audio+text"
}
df['assignment']= df["assignment"].replace(replace_dict)

replace_dict1 = {
    1: "Strongly Democratic",
    2: "Democratic",
    3: "Lean Democratic",
    4: "Lean Republican",
    5: "Republican",
    6: "Strongly Republican"
}
df['sa1']= df["sa1"].replace(replace_dict1)

# #replace_dict2 = {
#     1: "Strongly Conservative",
#     2: "Somewhat conservative",
#     3: "Moderate",
#     4: "Somewhat liberal",
#     5: "Strongly Liberal"
# }

# df['sa2']= df["sa2"].replace(replace_dict2)

replace_dict6 = {
    1: "Joseph Biden",
    2: "Donald Trump",
    3: "Other",
    4: "I'd rather not say",
    5: "I'm not eligible to vote"
}

df['sa6']= df["sa6"].replace(replace_dict6)

replace_dictct = {
    1: "Albania",
    2: "Brazil",
    3: "China",
    4: "Finland",
    5: "Germany",
    6: "India",
    7: "Israel",
    8: "Italy",
    9: "Korea",
    10: "Myanmar",
    11: "Peru",
    12: "Taiwan",
    13: "Trinidad and Tobago",
    14: "United Kingdom",
    15: "United States",
    16: "Venezuela, Bolivarian Republic of",
    17: "Vietnam"
}

df['Country of Birth']= df["Country of Birth"].replace(replace_dictct)


replace_dictW = {
    1: "Unemployed (and job seeking)",
    2: "Part-Time",
    3: "Full-Time",
    4: "Not in paid work (e.g. homemaker, retired or disabled)",
    5: "Due to start a new job within the next month",
    6: "Other"
}

df['Employment Status']= df["Employment Status"].replace(replace_dictW)

replace_dictsex = {
    1: "Female",
    2: "Male",
    3: "Prefer not to say"
}

df['Sex']= df["Sex"].replace(replace_dictsex)

df = df.rename(columns={
    'sa1': "political_leaning",
    'sa2': "social_leaning",
    'sa3': "economic_leaning",
    'sa4': "in_the_know",
    'sa5': "confi_in_media",
    'sa6': "voted_4",
    'sa7': "I_Know_abt_deepfakes"
})
```
